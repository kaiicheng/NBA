```python
class NormalEquation:
    def __init__(self, fit_intercept=True):
        self._fit_intercept = fit_intercept
    def fit(self, X_train, y_train):
        self._X_train = X_train.copy() # Not affect original data.
        self._y_train = y_train.copy() # Not affect original data.
        m = self._X_train.shape[0]
        
        # Determine add intercept or not.
        if self._fit_intercept:
            X0 = np.ones(m).reshape(-1, 1)
            self._X_train = np.concatenate([X0, self._X_train], axis=1) # axis=1 => row
        X_train_T = np.transpose(self._X_train)
        left_matrix = np.dot(X_train_T, self._X_train)  # (2, m)(m, 2) = (2, 2)
        right_matrix = np.dot(X_train_T, self._y_train) # (2, m)(m,) = (2,)
        left_matrix_inv = np.linalg.inv(left_matrix)    
        w = np.dot(left_matrix_inv, right_matrix)       # (2, 2)(2,) = (2,)
        self._w = w
        self.intercept_ = w[0]
        self.coef_ = w[1:]
    def predict(self, X):
        m = X.shape[0]
        if self._fit_intercept:
            X0 = np.ones(m).reshape(-1, 1)
            X = np.concatenate([X0, X], axis=1)
        y_pred = np.dot(X, self._w) # (m, 2)(2, ) = (m,)
        return y_pred
```

```python
h = NormalEquation()
h.fit(X_train, y_train)
```
